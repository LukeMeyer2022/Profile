```python
import pandas as pd

def describe_data(df, group_col, output_filename):
    # Grouping the DataFrame by the specified column
    grouped = df.groupby(group_col)
    
    # Applying the statistical summary to each group
    description = grouped.describe().loc[:, (slice(None), ['mean', 'std', 'min', '25%', '50%', '75%', 'max'])]
    
    # Save the DataFrame to a CSV file
    description.to_csv(output_filename)

    return description


group_col = "metabolicsyndrome" 
output_filename = 'stratified_summary_data.csv'
description = describe_data(df, group_col, output_filename)
description

```
