```python
# Import Packages and Import Raw Data
import pandas as pd
```

```python

data_path = "metabolic_syndrome.csv"
df = pd.read_csv(data_path)
df
```

```python
# Summary Statistics and Group By Metabolic Syndrome Status
import pandas as pd

def describe_data(df, group_col, output_filename):
    df_modified = df.drop(columns=['seqn'])
    grouped = df_modified.groupby(group_col)
    description = grouped.describe().round(3)
    description = description.loc[:, (slice(None), ['mean', 'std', 'min', '25%', '50%', '75%', 'max'])]
    description.to_csv(output_filename)

    return description

group_col = "metabolicsyndrome" 
output_filename = 'stratified_summary_data_status.csv'
description_data_status = describe_data(df, group_col, output_filename)
description_data_status
```
